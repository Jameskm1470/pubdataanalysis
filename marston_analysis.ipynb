```python
import requests
import json

def reverse_geocode(lat, lon):
    # Step 1: Make a request to the OpenStreetMap Nominatim API
    url = f'https://nominatim.openstreetmap.org/reverse?lat={lat}&lon={lon}&format=json'
    response = requests.get(url)

    # Step 2: Parse the response
    if response.status_code == 200:
        data = json.loads(response.text)
        #print(data)
        if 'address' in data:
            # Step 3: Extract the region name if available
            address_info = data['address']
            if 'state' in address_info:
                 if address_info['state']=='England':
                        if 'postcode' in address_info:
                            region = address_info['postcode']
                            return region
                 elif address_info['state']=='Cymru / Wales':
                    if 'postcode' in address_info:  
                        region='Wales'
                        return region
                 elif address_info['state']=='Alba / Scotland':
                    if 'postcode' in address_info:  
                        region='Scotland'
                        return region
                 elif address_info['state']=='Northern Ireland / Tuaisceart ireann':
                    if 'postcode' in address_info:  
                        region='Northern Ireland'
                        return region
def get_region_name(postcode):
    url = f"https://api.postcodes.io/postcodes/{postcode}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        region = data['result']['region']
        return region
    else:
        return None
        
x=[]    
for index, row in unique_addresses.iterrows():
    y=reverse_geocode(row['latitude'], row['longitude'])
    x.append(y)
region_name=[]
for pc in x:
          # Replace with the desired postcode
    if pc != 'Wales' and pc != 'Scotland' and pc != 'Northern Ireland':
        y = get_region_name(pc)
        print(y)
        region_name.append(y)
    else:
        region_name.append(pc)



```

```python
import numpy as np
# Set the figure size
plt.figure(figsize=(10, 6))

# Set the bar width
bar_width = 0.35

# Set the x positions of the bars
x = np.arange(len(df_filtered['State']))

# Plot the grouped bar chart
plt.bar(x - bar_width/2, df_filtered['pub_count'], width=bar_width, color='blue', alpha=0.7, label='Number of Pubs')
plt.bar(x + bar_width/2, df_filtered['Density'], width=bar_width, color='orange', alpha=0.7, label='Population Density')

# Add labels and title
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Number of Pubs and Population Density by Region')

# Add a legend
plt.legend()

# Set the x-axis ticks and labels
plt.xticks(x, df_filtered['State'])

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
